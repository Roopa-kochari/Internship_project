```python
from sklearn.preprocessing import OneHotEncoder
categorical_columns = ['Airline', 'Source', 'Destination']
encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
encoded_array = encoder.fit_transform(df[categorical_columns])
encoded_cols = encoder.get_feature_names_out(categorical_columns)
encoded_df = pd.DataFrame(encoded_array, columns=encoded_cols, index=df.index)
df_final = pd.concat([df.drop(categorical_columns, axis=1), encoded_df], axis=1)
```

```python
df_final['Date_of_Journey'].unique()
```

```python
df_final['Date_of_Journey'] = pd.to_datetime(df_final['Date_of_Journey'],dayfirst=True, errors='coerce')
df_final['Journey_Day'] = df_final['Date_of_Journey'].dt.day
df_final['Journey_Month'] = df_final['Date_of_Journey'].dt.month
df_final['Journey_Year'] = df_final['Date_of_Journey'].dt.year
```

```python
df_final.head(2)
```

```python
df_final.drop(columns=['Date_of_Journey'],axis=1,inplace=True)
```

```python
df_final.head(2)
```

```python
df_final['Duration'].unique()
```

```python
df_final['Duration_Minutes'] = df_final['Duration'].apply(
    lambda x: int(x.split('h')[0].strip()) * 60 + int(x.split('h')[1].replace('m','').strip()) if 'h' in x and 'm' in x 
    else int(x.split('h')[0].strip()) * 60 if 'h' in x 
    else int(x.replace('m','').strip())
)
```

```python
df_final.head(2)
```

```python
df_final.drop(columns=['Duration'],inplace=True,axis=1)
```

```python
df_final.head(2)
```

```python
df_final['Dep_Time'].unique()
```

```python
df_final['Dep_hour'] = df_final['Dep_Time'].str.split(':').str[0].astype(int)
df_final['Dep_minute'] = df_final['Dep_Time'].str.split(':').str[1].astype(int)
```

```python
df_final.head(2)
```

```python
df_final.drop(columns=['Dep_Time'],axis=1,inplace=True)
```

```python
df_final.head(2)
```

```python
df_final['Arrival_Time'].unique()
```

```python
def extract_hour_min(time_str):
    main = time_str.split()[0]
    if ':' in main:
        h, m = main.split(':')
    else:
        h, m = main[:2], main[2:4]
    return int(h), int(m)

df_final[['Arrival_hour','Arrival_minute']] = df_final['Arrival_Time'].apply(
    extract_hour_min
).apply(pd.Series)
```

```python
df_final.head(2)
```

```python
df_final.drop(columns=['Arrival_Time'],inplace=True,axis=1)
```

```python
df_final.head(2)
```

```python
print(df_final['Total_Stops'].unique())
```

```python
df_final =df_final.dropna(subset=['Total_Stops'])
from sklearn.preprocessing import OrdinalEncoder
encoder = OrdinalEncoder(categories=[['non-stop', '1 stop', '2 stops', '3 stops', '4 stops', 'missing']])
df_final['TotalStops'] = encoder.fit_transform(df_final[['Total_Stops']])
df_final['TotalStops'] = df_final['TotalStops'].astype(int)
```

```python
df_final.head(2)
```

```python
df_final.drop(columns=['Total_Stops'],inplace=True,axis=1)
```

```python
df_final.head(2)
```

MODEL BULDING

```python
X=df_final.drop('Price',axis=1)
y=df_final[['Price']]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
LR_model=LinearRegression()
LR_model.fit(X_train,y_train)
```

```python
y_pred=LR_model.predict(X_test)
y_pred
```

```python
LR_model.coef_
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
LR_r2_score = r2_score(y_test,y_pred)
LR_mean_squared_error = mean_squared_error(y_test,y_pred)
LR_mean_absolute_error = mean_absolute_error(y_test,y_pred)
```

```python
LR_result= pd.DataFrame({ LR_r2_score,'LinearRegression'}).transpose()
LR_result.columns=['r2_score','Model']
LR_result
```
